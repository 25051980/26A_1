# Tarefa - Álgebra Linear

A célula abaixo gera dados de notas de duas provas, p1 e p2, conforme em aula, mas com mais alunos.

## 1. Reduzindo correlações

a) Já vimos que expressar os dados através da média e diferença é uma transformação linear. Obtenha a média e a diferença através de uma multiplicação matricial, conforme visto em aula.

b) Calcule a correlação entre os dados da **p1** e **p2**

c) Calcule a correlação entre os dados transformados (média e diferença)

d) Comente as correlações observadas.

```python
import pandas as pd
import seaborn as sns
import numpy as np
import matplotlib.pyplot as plt

%matplotlib inline

# Gerando os dados
np.random.seed(123)

p1 = np.random.random(50) *10
p1[p1>10] = 10
p1[p1<0] = 0

p2 = p1 + np.random.normal(0,1,50)
p2[p2>10] = 10
p2[p2<0] = 0

df = pd.DataFrame({'p1': p1, 'p2': p2})
```

## 2. Soma e diferença

Realize agora uma transformação semelhante, mas que ao invés da média, dê a soma da **p1** e **p2**. Construa o gráfico e conclua que diferenças visuais você observa.

Este código realiza uma transformação nos dados calculando a soma das variáveis `p1` e `p2`, adicionando-a como uma nova coluna chamada `soma` no DataFrame. Em seguida, o gráfico compara a relação original entre `p1` e `p2` (em azul) com a relação transformada entre `p1` e a soma (`p1 + p2`, em laranja). Observa-se que a transformação desloca os valores para cima, ampliando visualmente a correlação e intensificando a relação linear entre as variáveis, indicando que a soma pode introduzir maior dependência entre as variáveis.


```python
# Calculando a soma de p1 e p2
df['soma'] = df['p1'] + df['p2']

# Gerando o gráfico para visualizar a soma
plt.figure(figsize=(10, 6))
sns.scatterplot(x=df['p1'], y=df['p2'], label='Original (p1 vs p2)', color='blue')
sns.scatterplot(x=df['p1'], y=df['soma'], label='Transformado (p1 vs soma)', color='orange')
plt.title('Comparação entre p1, p2 e Soma (Transformação)')
plt.xlabel('p1')
plt.ylabel('p2 e Soma')
plt.legend()
plt.grid(alpha=0.3)
plt.show()

# Comentando as diferenças
print("Diferenças visuais:")
print("1. A soma (linha laranja) desloca os pontos para valores maiores em comparação com os valores originais (linha azul).")
print("2. Isso ocorre porque a soma aumenta a correlação e amplifica a relação linear entre p1 e p2.")
```

o gráfico mostra a comparação entre os dados originais (`p1` versus `p2`, representados em azul) e os dados transformados (`p1` versus a soma `p1 + p2`, representados em laranja). Observa-se que a transformação pela soma desloca os pontos para valores mais altos, indicando um aumento na correlação entre as variáveis. Essa transformação amplifica a relação linear entre `p1` e `p2`, tornando os dados mais dependentes visualmente e influenciando a distribuição das observações no gráfico.
